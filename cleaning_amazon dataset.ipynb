```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Amazon Sale Report.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'],format='%m-%d-%y')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['New','PendingS'],inplace=True)
```

```python
df['fulfilled-by'].fillna('Unknown',inplace=True)
```

```python
df['Amount']=df['Amount'].fillna(df['Amount'].mean())
```

```python
df['Amount'].dtype
```

```python
df['Amount'].isnull().sum()
```

```python
df['Amount'].head(10)
```

```python
df['Amount'].describe()
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].median())
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['ship-city','ship-state','ship-postal-code','ship-country'])
```

```python
df.drop(columns=['currency'],inplace=True)
```

```python
df['ship-state'].unique()
```

```python
df['ship-city'].unique()
```

```python
df['ship-state']=df['ship-state'].str.upper()
```

```python
df['ship-state']=df['ship-state'].str.strip()
```

```python
df['ship-state']=df['ship-state'].replace({'ORISSA':'ODISHA','PB':'PUNJAB','PUNJAB/MOHALI/ZIRAKPUR':'PUNJAB','RJ':'RAJASTHAN','RAJSTHAN':'RAJASTHAN',
  'RAJSTHAN':'RAJASTHAN','NEW DELHI':'DELHI','NL':'NAGALAND','AR':'ARUNACHAL PRADESH','PONDICHERRY':'PUDUCHERRY'})                                         



                                           
```

Top Sales by State 

```python
top_state_sales=df.groupby('ship-state')['Amount'].sum().sort()_values(ascending=False)top_state_sales.head(15)
```

```python
top_state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)
top_state_sales.head(60)
```
